```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate
import cv2
```

```python
W9A = cv2.imread('W9A.jpg')
D3A = cv2.imread('D3A.jpg')
P4A = cv2.imread('P4A.jpg')
E5A = cv2.imread('E5A.jpg')
T6A = cv2.imread('T6A.jpg')
G7D = cv2.imread('G7D.jpg')
G7R = cv2.imread('G7R.jpg')
T8A = cv2.imread('T8A.jpg')
Y10A = cv2.imread('Y10A.jpg')
```

```python
columns = 3
rows = 4
fig = plt.figure(figsize =(15,15))
fig.subplots_adjust(hspace=0.1,wspace =0.05)
fig.add_subplot(3,3,1)
plt.imshow(cv2.cvtColor(D3A, cv2.COLOR_BGR2RGB),aspect='auto')
plt.axis('off')
fig.add_subplot(3,3,2)
plt.imshow(cv2.cvtColor(P4A, cv2.COLOR_BGR2RGB),aspect='auto')
plt.axis('off')
fig.add_subplot(3,3,3)
plt.imshow(cv2.cvtColor(E5A, cv2.COLOR_BGR2RGB),aspect='auto')
plt.axis('off')
fig.add_subplot(3,3,4)
plt.imshow(cv2.cvtColor(T6A, cv2.COLOR_BGR2RGB),aspect='auto')
plt.axis('off')
fig.add_subplot(3,3,5)
plt.imshow(cv2.cvtColor(G7D, cv2.COLOR_BGR2RGB),aspect='auto')
plt.axis('off')
fig.add_subplot(3,3,6)
plt.imshow(cv2.cvtColor(G7R, cv2.COLOR_BGR2RGB),aspect='auto')
plt.axis('off')
fig.add_subplot(3,3,7)
plt.imshow(cv2.cvtColor(T8A, cv2.COLOR_BGR2RGB),aspect='auto')
plt.axis('off')
fig.add_subplot(3,3,8)
plt.imshow(cv2.cvtColor(W9A, cv2.COLOR_BGR2RGB),aspect='auto')
plt.axis('off')
fig.add_subplot(3,3,9)
plt.imshow(cv2.cvtColor(Y10A, cv2.COLOR_BGR2RGB),aspect='auto')
plt.axis('off')
#fig.tight_layout()
plt.savefig('present2.jpg',bbox_inches='tight',dpi =500)
plt.show()
```
